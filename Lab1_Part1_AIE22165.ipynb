```python
#Q2
import random
import time
import matplotlib.pyplot as plt

# Function to generate an array of random integers
def generate_random_array(size):
    return [random.randint(1, 1000) for _ in range(size)]

# Linear search algorithm
def linear_search(arr, key):
    for i in range(len(arr)):
        if arr[i] == key:
            return i
    return -1

# Binary search algorithm
def binary_search(arr, key):
    low = 0
    high = len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == key:
            return mid
        elif arr[mid] < key:
            low = mid + 1
        else:
            high = mid - 1
    return -1

# Generate random array of 10,000 elements
arr = generate_random_array(10000)

# Perform searches and plot time taken
searches = 5
search_key_times_linear = []
search_key_times_binary = []

for _ in range(searches):
    search_key = int(input("Enter search key: "))

    # Time linear search
    start_time = time.time()
    linear_search(arr, search_key)
    end_time = time.time()
    search_key_times_linear.append(end_time - start_time)

    # Time binary search
    start_time = time.time()
    binary_search(arr, search_key)
    end_time = time.time()
    search_key_times_binary.append(end_time - start_time)

# Plotting
plt.plot(range(1, searches + 1), search_key_times_linear, label='Linear Search')
plt.plot(range(1, searches + 1), search_key_times_binary, label='Binary Search')
plt.xlabel('Search')
plt.ylabel('Time Taken (seconds)')
plt.title('Time taken by Linear and Binary Search')
plt.legend()
plt.show()
```

```python
#Q3

def convert_to_int(string):
    if not string:
        return 0
    else:
        digit = int(string[0])
        remaining_string=string[1:]
        return digit*(10 ** len(remaining_string)) + convert_to_int(remaining_string)
    
num_as_str=input("Enter the number to be converted into integer: ")
convert_to_int(num_as_str)
```

```python
#Q4
def reverse(string):
    if not string:
        return ""
    else:
        return string[len(string)-1] + reverse(string[:len(string)-1])
word = input("Enter the word to reverse: ")
reverse(word)
```

```python
#Q5
def palindrome(string):
    if not string:
        return ""
    else:
        return string[len(string)-1] + palindrome(string[:len(string)-1])
word = input("Enter the word to reverse: ")
reverse=palindrome(word)
if word==reverse:
    print("The given word is a palindrome")
else:
    print("The given word is not a palindrome")
```
